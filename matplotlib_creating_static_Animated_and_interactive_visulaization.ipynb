```python
import matplotlib.pyplot as plt
import numpy as np # it is used to create the array of random number
```

```python
x = np.linspace(0,10,40)
# this linspace number will generate the 40 number between 0 and 10
y = x**2
print(x)
print(y)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,"r")
plt.xlabel("Number range")
plt.ylabel("squared output")
plt.title("string titile here")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.5,0.5,0.4,0.4])
```

```python
x = np.linspace(0,10,1000)
y = np.sin(x)
plt.plot(x,y,"b")
plt.show()
```

bar chart

```python
x = [5,25,50,20]
plt.bar(range(len(x)),x)
```

```python
plt.figure(figsize=(7,7))
x = [35,25,20,20]
labels = ["computer","Electronics","Mechnical","chemical"]
plt.pie(x,labels = labels)
plt.show()
```
